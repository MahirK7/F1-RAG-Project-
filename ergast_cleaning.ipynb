```python
# ================================
# FastF1 Cleaning (Race Results)
# ================================

import fastf1
import pandas as pd
import os

# Enable cache
fastf1.Cache.enable_cache("fastf1_cache")

# Load the Australia 2023 race session
session = fastf1.get_session(2023, "Australia", "R")
session.load()

# Get classification
results = session.results

# Create processed folder
os.makedirs("processed", exist_ok=True)

# Save full results to CSV
csv_path = "processed/ergast_clean.csv"
results.to_csv(csv_path, index=False)
print(f"✅ Race results saved to {csv_path}")

# Preview top 10 in console
print(results[["Position", "Abbreviation", "FullName", "TeamName", "Status"]].head(10))

# Convert into text for RAG
ergast_text = "\n".join([
    f"P{row.Position}: {row.FullName} ({row.Abbreviation}) - {row.TeamName}, Status: {row.Status}"
    for _, row in results.iterrows()
])

# Save as txt for vectorstore
txt_path = "processed/ergast_australia2023.txt"
with open(txt_path, "w", encoding="utf-8") as f:
    f.write(ergast_text)

print(f"✅ Ergast-style text exported to {txt_path}")
```
